```python
import joblib
import pandas as pd

# Cargar el pipeline completo
pipeline = joblib.load("pipeline_random_forest.pkl")

# Cargar el DataFrame limpio
df_clean = pd.read_csv("df_clean.csv")

# Aplicar el pipeline al DataFrame limpio y obtener predicciones
y_pred = pipeline.predict(df_clean)

# Agregar las predicciones como una nueva columna
df_clean["Predicciones"] = y_pred

# Verificar las primeras filas para confirmar las predicciones
print(df_clean.head())
```

```python
import matplotlib.pyplot as plt

# Gr√°fico de dispersi√≥n para comparar los valores de 'price' con las predicciones
plt.figure(figsize=(10, 6))
plt.scatter(df_clean['price'], df_clean['Predicciones'], alpha=0.5, color='skyblue')
plt.plot([df_clean['price'].min(), df_clean['price'].max()],
        [df_clean['price'].min(), df_clean['price'].max()], 'k--', lw=2)
plt.xlabel("Valores Reales (price)")
plt.ylabel("Predicciones")
plt.title("Comparaci√≥n de Valores Reales (price) vs Predicciones")
plt.show()
```

```python
# Gr√°fico de barras para las primeras 60 observaciones
n = 60  # N√∫mero de observaciones a mostrar
plt.figure(figsize=(14, 6))
df_clean[['price', 'Predicciones']].head(n).plot(kind='bar', figsize=(14, 6), color=['blue', 'orange'])
plt.title(f"Comparaci√≥n de los primeros {n} Valores Reales (price) vs Predicciones")
plt.xlabel("√çndice")
plt.ylabel("Valores")
plt.xticks(rotation=45)
plt.show()
```

```python
df_clean.to_csv("df_clean_con_predicciones.csv", index=False)
```
